```python
#this function prints the grid and all instances of objects
def makeGrid(rows,cols,monsters,player,eggs,door):
    eggCoords=[egg.coords for egg in eggs]
    for y in range(rows):
        print(' ---'*(cols))
        for x in range(cols):
            monster_created=False
            for monster in monsters:
                if monster.coords == [x,y] and x==cols-1:
                    print('| m |',end='')
                    monster_created=True
                elif monster.coords == [x,y]:
                    print('| m ',end='')
                    monster_created=True
            if [x,y] in eggCoords and x==cols-1 and monster_created==False:
                print('| e |',end='')
            elif [x,y] in eggCoords and monster_created==False:
                print('| e ',end='')
            elif player.coords == [x,y] and x==cols-1 and monster_created==False:
                print('| p |',end='')
            elif player.coords == [x,y] and monster_created==False:
                print('| p ',end='')
            elif door.coords == [x,y] and x==cols-1 and monster_created==False:
                print('| d |',end='')
            elif door.coords == [x,y] and monster_created==False:
                print('| d ',end='')
            elif x ==cols-1 and monster_created==False:
                print('|   |',end='')
            elif monster_created==False:
                print('|   ',end='')
        print()
        if y == rows-1:
            print(' ---'*(cols))
            
```

```python
import random 
from IPython.display import clear_output

def outofbounds(obj,col,row):
    if 0<obj.coords[0]<col and 0<obj.coords[1]<row:
        return True
    else:
        return False

class Monster():
    def __init__(self,coords):
        self.coords=coords
    def initCoords(self,rows,cols):
        self.coords=[random.randint(0,cols-1),random.randint(0,rows-1)]
    def move(self,step=1,x=0,y=0):
        delta=random.choice([-1*step,step])
        for i in range(delta):
            if random.choice([True,False]):
                x+=(delta)/abs(delta)
            else:
                y+=(delta)/abs(delta)
        self.coords[0]+=x
        self.coords[1]+=y
    
class Player():
    def __init__(self,coords,eggs=0):
        self.coords=coords
        self.eggs=eggs
    def initCoords(self,rows,cols,monsters):
        #initialize coordinates
        self.coords=[random.randint(0,cols-1),random.randint(0,rows-1)]
        #makes sure have different coords to monster
        for monster in monsters:
            while monster.coords==self.coords:
                self.coords=[random.randint(0,cols-1),random.randint(0,rows-1)]
    def move(self,cols,rows):
        u_input=input('Where would you like to move? ')
        if u_input=='up' or u_input=='u':
            self.coords[1]-=1
        elif u_input=='down' or u_input=='d':
            self.coords[1]+=1
        elif u_input=='left'or u_input=='l':
            self.coords[0]-=1
        elif u_input=='right' or u_input=='r':
            self.coords[0]+=1
    
        
class Egg():
    def __init__(self,coords):
        self.coords=coords
    def initCoords(self,rows,cols):
        self.coords=[random.randint(0,cols-1),random.randint(0,rows-1)]        
class Door():
    def __init__(self,coords):
        self.coords=coords
    def initCoords(self,rows,cols):
        self.coords=[random.randint(0,cols-1),random.randint(0,rows-1)]
```

```python
#pickup eggs
def checkEgg(player,eggs):
    for egg in eggs:
        if player.coords== egg.coords:
            player.eggs+=1
            egg.coords=[-1,-1]
```

```python
def game_over(player,monsters,eggs_needed,door):
    for monster in monsters:
        if monster.coords==player.coords:
            return 1
    if player.coords==door.coords and player.eggs==eggs_needed:
        return 2
    else:
        return False


#handles player input
flag = False

levels=1

#main outer initializing loop
while True:
    #size of grid
    rows=5+(levels//2)
    cols=5+(levels//2)
    
    #objects for game
    monsters=[Monster([0,0]) for i in range(levels)]
    player=Player([1,1])
    eggs=[Egg([2,2]) for i in range(1+(levels//3))]
#     egg=Egg([2,2])
    door=Door([3,3])
    
    #randomly initiating coordinates for start
    for monster in monsters:
        monster.initCoords(rows,cols)
    player.initCoords(rows,cols,monsters)
    player.eggs=0
    for egg in eggs:
        egg.initCoords(rows,cols)
    door.initCoords(rows,cols)
               
               
    if flag==2:
        levels+=1
    #main game loop
    while True:
        clear_output()
        print("Eggs Collected: {}".format(player.eggs))
        #show grid:
        makeGrid(rows,cols,monsters,player,eggs,door)
        #move player:
        player.move(cols,rows)
#         while player.coords[0] not in range(cols-1) or player.coords[1] not in range(rows-1):
#             player.move(cols,rows)
        #move monsters:
        for monster in monsters:
            monster.move()
            while monster.coords[0] not in range(cols-1) or monster.coords[1] not in range(rows-1):
                if monster.coords[0]<0:
                    monster.move(0,2,0)
                elif monster.coords[0]>cols-1:
                    monster.move(0,-2,0)
                elif monster.coords[1]<0:
                    monster.move(0,0,2)
                elif monster.coords[1]>rows-1:
                    monster.move(0,0,-2)
                else:
                    break
        #check game_over
        flag=game_over(player,monsters,len(eggs),door)
        if flag==1:
            clear_output()
            print('You were eaten by the monster!')
            break
        elif flag==2:
            clear_output()
            print('You collected the eggs! ')
            break
        checkEgg(player,eggs)
    #see if they want to continue playing
    u_input=input('Would you like to play some more? ').lower()
    if u_input=='no' or u_input=='quit':
        break
            
```
